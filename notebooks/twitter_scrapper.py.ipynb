```python
from packages.scrapper import twitter_scrapper, helper
import pandas as pd
```

```python
tscrapper = twitter_scrapper.TwitterScrapper()
keywords = "machine learning jobs"
tweets = tscrapper.scrape(keywords)
```

```python
headers = ['Name', 'Screen Name', 'Text', 'Urls']
rows = []
for tweet in tweets:
    rows.append([tweet['name'], tweet['screen_name'], tweet['tweet_text'], tweet['urls']])
helper.display_table((headers, rows))
```
